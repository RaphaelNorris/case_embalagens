```python
import pandas as pd
import numpy as np
from pathlib import Path

# =========================
# Tipagem por prefixo (sem coerce, sem '.0')
# =========================

def _coerce_boolean(s: pd.Series) -> pd.Series:
    s = s.copy()
    s = s.fillna(False)
    s = s.replace({
        1: True, 0: False,
        '1': True, '0': False,
        'True': True, 'False': False,
        'true': True, 'false': False,
        'S': True, 'N': False,
        'Sim': True, 'Não': False,
        'NAO': False
    })
    return s.astype('boolean')


def _clean_id_like_values(s: pd.Series) -> pd.Series:
    """
    Remove sufixo '.0' de valores ID/CD que vieram como float.
    Exemplo: 49.0 -> '49'; 100.5 -> '100.5'
    """
    s = s.fillna('-1').astype(str)
    # Remove .0 apenas se for realmente decimal "inteira"
    s = s.str.replace(r'\.0$', '', regex=True)
    # Remover espaços e normalizar
    return s.str.strip()


def cast_series_by_prefix(s: pd.Series, prefix: str) -> pd.Series:
    s = s.copy()

    # IDs e textos
    if prefix in {"CD", "ID"}:
        s = _clean_id_like_values(s)
        return s

    elif prefix in {"TX", "ST"}:
        s = s.fillna('-1').astype(str).str.strip()
        return s

    # Valores numéricos (float)
    elif prefix == "VL":
        s = s.fillna(0)
        try:
            s = s.astype(float)
        except ValueError:
            s = pd.Series(
                [float(x) if str(x).replace('.', '', 1).isdigit() else 0 for x in s],
                index=s.index
            )
        return s

    # Quantidades (inteiros ou floats)
    elif prefix == "QT":
        s = s.fillna(0)
        try:
            s = s.astype(float)
        except ValueError:
            s = pd.Series(
                [float(x) if str(x).replace('.', '', 1).isdigit() else 0 for x in s],
                index=s.index
            )

        all_integers = np.isclose(s, np.round(s)).all()
        if all_integers:
            s = s.astype('Int64')
        return s

    # Booleanos
    elif prefix in {"FL", "BL"}:
        return _coerce_boolean(s)

    # Datas
    elif prefix == "DT":
        s = s.fillna('2999-01-01')
        return pd.to_datetime(s)

    return s


def apply_types_by_prefix(df: pd.DataFrame) -> pd.DataFrame:
    for col in df.columns:
        if "_" in col:
            prefix = col.split("_", 1)[0]
            df[col] = cast_series_by_prefix(df[col], prefix)
    return df


def rename_with_prefix(df: pd.DataFrame, mapping: dict) -> pd.DataFrame:
    def rename_col(c):
        if "_" in c:  # já tem prefixo
            return c
        return f"{mapping[c]}_{c}" if c in mapping else c
    return df.rename(columns=rename_col)


def process_table(filepath: str, mapping: dict):
    df = pd.read_parquet(filepath)
    df = rename_with_prefix(df, mapping)
    df = apply_types_by_prefix(df)

    output = Path(filepath).with_name("typed_" + Path(filepath).name)
    df.to_parquet(output)
    print(f" Processado: {filepath} → {output}")

# Itens
process_table('../../data/02 - trusted/parquet/tb_itens.parquet', prefixos_itens)

# Pedidos
process_table('../../data/02 - trusted/parquet/tb_pedidos.parquet', prefixos_pedidos)

# Clientes
process_table('../../data/02 - trusted/parquet/tb_clientes.parquet', prefixos_clientes)

# Facas
process_table('../../data/02 - trusted/parquet/tb_facas.parquet', prefixos_facas)

# Máquinas
process_table('../../data/02 - trusted/parquet/tb_maquina.parquet', prefixos_maquinas)

# Paradas
process_table('../../data/02 - trusted/parquet/tb_paradas.parquet', prefixos_paradas)

# Tarefas de conversão (tarefcon)
process_table('../../data/02 - trusted/parquet/tb_tarefcon.parquet', prefixos_tarefcon)
```
