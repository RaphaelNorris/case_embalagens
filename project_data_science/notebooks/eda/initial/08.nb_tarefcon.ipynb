```python
import pandas as pd

df_tarefcon =  pd.read_parquet('../../data/02 - trusted/parquet/tb_tarefcon.parquet')
```

```python
df_tarefcon = df_tarefcon[sorted(df_tarefcon.columns)]

df_tarefcon['CD_CODIGOPARADAOUCONV'].fillna(str(0),inplace=True)
df_tarefcon['CD_CODIGOPARADAOUCONV'] = df_tarefcon['CD_CODIGOPARADAOUCONV'].astype(float)
df_tarefcon['CD_CODIGOPARADAOUCONV'] = df_tarefcon['CD_CODIGOPARADAOUCONV'].astype('Int64')
df_tarefcon[['CD_CODIGOPARADAOUCONV', 'CD_FACA', 'CD_MAQUINA', 'CD_OP',
       'CD_OPONDULADA', 'CD_ORIGEMREGISTRO', 'CD_TURMA']].fillna(str(0),inplace=True)
```
