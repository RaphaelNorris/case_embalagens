### Tarefcon

```python
df_tarefcon_trusted = pd.read_sql("""
    SELECT
        ID_MAQUINA,
        FL_PARADA,
        CD_PARADAOUCONV,
        TX_TURMA,
        TX_OP,
        ID_PEDIDO,
        ID_ITEM,
        VL_REPROGRAMACAO,
        QT_ARRANJO,
        VL_GRAMATURA,
        QT_PROGRAMADA,
        VL_CHAPASALIMENTADAS,
        QT_PRODUZIDA,
        QT_AJUSTE,
        VL_DURACAOPREVISTA,
        DT_INICIO,
        DT_FIM,
        DT_DIADATURMA,
        ID_CLIENTE,
        ID_USUARIO,
        VL_ORIGEMREGISTRO,
        TX_DESCORIGEMREGISTRO,
        FL_SKIPFEED,
        TX_OPONDULADA,
        VL_DURACAO,
        CD_FACA
    FROM TAREFCON
    WHERE DT_DIADATURMA >= TO_DATE('2022-01-01', 'YYYY-MM-DD')""", con=conn)
```

```python
df_tarefcon = df_tarefcon_trusted.drop_duplicates().reset_index(drop=True)

df_tarefcon = clean_tarefcon(df_tarefcon)
df_tarefcon = df_tarefcon[sorted(df_tarefcon.columns)]
# df_tarefcon.to_parquet('./../../../data/trusted/parquet/tb_tarefcon.parquet')
df_tarefcon = pd.read_parquet('./../../../data/trusted/parquet/tb_tarefcon_trusted.parquet')
```
